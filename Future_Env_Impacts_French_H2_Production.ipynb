### Generate Prospective Database

```python
from datapackage import Package

# import gc
# gc.collect()


fp = "datapackage_attributional.json" # Or adapt with your the file path

h2 = Package(fp)
h2.resource_names

scenario = "N03_UB-H2" # should be written the same as in scenario data, configuration and .json file and 
model = "TIAM-UCL"
pathway = "SSP2-RCP45"

external_scenario = [
    {"scenario": scenario , "data": h2}
]

for year in [2025] :#[2025, 2030, 2035]: # The more the number of years the more the memory needed to fully complete the loop
    print(year)

    scenarios = [
        {"model": model, "pathway":pathway, "year": year, "external scenarios": external_scenario},
    ]

    ndb = NewDatabase(
        scenarios = scenarios,        
        source_db="ecoinvent-3.11-cutoff",
        source_version="3.11",
        #key='xxx', # The key is mandatory to use Premise
        quiet = True,
    )
    #ndb.generate_scenario_report()
    
    ndb.update(["electricity", # integrate IAM projections
                "external"]) # integrate custom scenario projections on top of it
    ndb.write_db_to_brightway(scenario+'_'+model+'_'+pathway+'_'+str(year)+'bis')

# gc.collect()
```
